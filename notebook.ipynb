```python
import pandas as pd
import numpy as np
import os

train = pd.read_csv('training.csv', sep=';')

train = train.dropna(how='any')

train.head
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

ohe = preprocessing.OneHotEncoder(categories = "auto")

categorical_feature_mask = train.dtypes==object
categorical_cols = train.columns[categorical_feature_mask].tolist()

train[categorical_cols] = train[categorical_cols].apply(lambda col: le.fit_transform(col))

X = train.iloc[:,train.columns != 'classLabel']

X = ohe.fit_transform(X)

y = train['classLabel']

print(X.shape)
```

```python
from sklearn.linear_model import LogisticRegression  

logreg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr').fit(X, y)

logreg
```

```python
test = pd.read_csv('validation.csv', sep=';')

test = test.dropna(how='any')

test.head
```

```python
categorical_feature_mask = test.dtypes==object
categorical_cols = test.columns[categorical_feature_mask].tolist()

test[categorical_cols] = test[categorical_cols].apply(lambda col: le.fit_transform(col))

X_test = test.iloc[:,test.columns != 'classLabel']

X_test = ohe.fit_transform(X_test)

print(X_test.shape)
```

```python
# Solve mismatch between features of two datasets
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```
